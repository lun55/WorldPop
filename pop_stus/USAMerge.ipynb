```python
import pandas as pd
from pathlib import Path

years = [2021, 2022, 2023]
days = ["night", "oneday"]

# 作为“键”的列
key_cols = [
    'GID_2',
    'Province',
    'City',
    'Gender',
    'Age',
    'Noise_Threshold'
]

base_dir = Path(r"E:\Code\fastServer\机场噪声\2021-2022美国")

for year in years:
    for day in days:
        try:
            data_dir = base_dir / str(year) / day

            file_e = data_dir / "Aleutians_West_E_affected_population.csv"
            file_w = data_dir / "Aleutians_West_W_affected_population.csv"
            file_out = data_dir / "Aleutians_West_affected_population.csv"

            # 文件存在性检查
            if not file_e.exists() or not file_w.exists():
                print(f"⚠️ 跳过：{year}/{day}（缺少 E 或 W 文件）")
                continue

            # 读取数据
            df_e = pd.read_csv(file_e)
            df_w = pd.read_csv(file_w)

            # 合并并累加
            df_merge = (
                pd.concat([df_e, df_w], ignore_index=True)
                  .groupby(key_cols, as_index=False)['Affected_Pop']
                  .sum()
            )

            # 输出
            df_merge.to_csv(file_out, index=False)

            print(f"✅ 完成：{year}/{day} → {file_out.name}")

        except Exception as e:
            print(f"❌ 出错：{year}/{day}\n    错误信息：{e}")
```
